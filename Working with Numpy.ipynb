```python
import numpy as np
```

```python
python_list=[1,2,3,4,5]
print(f"Python List: {python_list}")

numpy_list=np.array([1,2,3,4,5])
print(f"Numpay list: {numpy_list}")

print("Added 10 in each Element:",[x+10 for x in python_list])

print("Added 10 in numpy list",numpy_list+10)
```

```python
import time

size=1000000
python_list = list(range(size))

numpy_list=np.arange(size)

start=time.time()
result_list=[x+1 for x in python_list]
time_for_python_list=time.time()-start
print(f"The operation was done in: {time_for_python_list:.4f}s")

start=time.time()
result_numpy_list=numpy_list+1
time_for_numpy_list=time.time()-start
print(f"The operation was completed in: {time_for_numpy_list:.4f}s ")
print(f"Numpy is: {time_for_python_list/time_for_numpy_list:.4f}s Faster")
```
